```python
from collections import defaultdict

lines = [line.strip() for line in open('inputs/day19.txt')]
start = lines[-1]
replacements = lines[:-2]

reps = defaultdict(list)
for line in replacements: 
    key, val = line.split(' => ')
    reps[key].append(val)
```

```python
def findall(string, value):
    current_search = 0
    while True: 
        index = string[current_search:].find(value)
        if index == -1:
            break
            
        
        yield current_search+index
        current_search = current_search + index + len(value)

[x for x in findall('heyhey', 'ey')]

def get_all_mutations(start, replacements):
    all_sentences = list()
    for before in replacements: 
        for after in replacements[before]:
            
            for index in findall(start, before):
                newstring = start[:index] + after + start[index+len(before):]
                all_sentences.append(newstring)
    return all_sentences

res = get_all_mutations(start, reps)

print('Part 1', len(set(res)))
```

```python
def get_max_overlap(a, b):
    for i in range(min(len(a), len(b))):
        if a[i] != b[i]:
            return i
    return min(len(a), len(b))

molecules = ['e']

iterations = 0
while start not in molecules: 
    print('iteration', iterations)
    nextmolecules = list()
    for m in molecules: 
        nextmolecules.extend(get_all_mutations(m, reps))
    molecules = list(set(nextmolecules))
    molecules = [(get_max_overlap(start, x), x) for x in molecules]
    molecules.sort(reverse=True)
    molecules = [x[1] for x in molecules[:400]]

    iterations += 1
    
print("Part 2:", iterations)
```
