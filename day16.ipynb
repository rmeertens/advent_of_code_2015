```python
lines = [line.strip() for line in open('inputs/day16.txt')]
```

```python
lines
```

```python
import re

sues = dict()

for line in lines: 
    groups = re.match(r'^Sue (.*): (.*): (.*), (.*): (.*), (.*): (.*)$', line)
    number = groups.group(1)    
    propname1, propval1, propname2, propval2, propname3, propval3 = [groups.group(x) for x in range(2, 8)]

    sues[number] = dict()
    sues[number][propname1] = propval1
    sues[number][propname2] = propval2
    sues[number][propname3] = propval3
```

```python
request = """children: 3
cats: 7
samoyeds: 2
pomeranians: 3
akitas: 0
vizslas: 0
goldfish: 5
trees: 3
cars: 2
perfumes: 1"""
MFCSAM = dict()
for line in request.split("\n"):
    name, val = line.split(": ")
    MFCSAM[name] = val
```

```python
for sue in sues.keys(): 
    matches = True
    for key in MFCSAM: 
        if key not in sues[sue]:
            continue
        else: 
            local_ok = MFCSAM[key] == sues[sue][key]
            matches = matches and local_ok
    if matches: 
        print('Part 1', sue)
```

```python
for sue in sues.keys(): 
    matches = True
    for key in MFCSAM: 
        if key not in sues[sue]:
            continue
        else: 
            if key in ['cats', 'trees']: 
                local_ok = sues[sue][key] > MFCSAM[key]
            elif key in ['pomeranians', 'goldfish']:
                local_ok = sues[sue][key] < MFCSAM[key]
            else:
                local_ok = MFCSAM[key] == sues[sue][key]
            matches = matches and local_ok
    if matches: 
        print('Part 2', sue)
```
